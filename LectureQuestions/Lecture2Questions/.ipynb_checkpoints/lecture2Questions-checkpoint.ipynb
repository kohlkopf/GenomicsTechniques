2. *What is paired end sequencing? Describe why you would perform paired-end sequencing.*
    + The sequencing is duplicated but in the opposite direction. If the first read was from the 5' to the 3' end, the other read will be from the 3' to the 5' end. THis is accomplished using a a primer that attaches to the 3' end. Paired end reads are much easier to align.

    2. *describe one solution*
        + Here are several tips to mitigate hopping recommended by Illumina:
            + Remove free adaptors from library preps
            + Store libraries individually at -20°C
            + Pool libraries prior to sequencing
            + Use unique dual indexing pooling combinations (unique i5 and i7 indexes)
        + A solution being implemented by Illumina is to use dual indexes. These will enable researchers to remove unexpected combinations and only focus on the ‘true’ data with correct index combinations.
    
4. *What are the advantages to Pac Bio? What might be a disadvantage?*
    + The PacBio workflow produces very long reads, runs massively parallel and is high throughput, and has a random error mode. This long read sequencer may not be the best for short sequencing needs. 
5. *Why did Illumina switch to binning quality scores?*
    + Illumina bins quality scores to save memory. The ASCII encoding for each read took up too much space.
6. *Name two methods of sequencing other than sequencing by synthesis and briefly describe how they work.*
    + Roche/454--pyrosequencing
        + The process begins with a sample of Double-stranded DNA is fragmented with a restriction enzyme.
        + Adapters are attached to the DNA fragments.
        + Tiny resin beads are added to the mix.
        + The beads contain complementary sequences to the adapters allowing each fragment to bind to a bead.
        + When the fragments attach to the beads the DNA denatures.
        + The fragments of DNA are then copied numerous times on each bead by PCR.
        + The beads are then filtered to remove any that without a fragment or contain more than one DNA fragment.
        + The remaining beads are put into wells on a sequencing plate (one bead per well) along with enzyme beads that contain the DNA polymerase and primer needed for the sequencing reaction.
        + The polymerase enzyme and primer attach to the DNA fragments on the beads.
        + Nucleotide bases are added to the wells one base at a time. Optics record the fluoresence.
        + The intensity of the light corresponds to the number of nucleotides of the same type that have been incorporated. 
        + By plotting this pattern of light intensity on a graph, the sequence of the original piece of DNA can be decoded.
    + Ion Torrent
        + The process begins with a sample of Double-stranded DNA is fragmented with a restriction enzyme.
        + Adapters are attached to the DNA fragments.
        + Tiny resin beads are added to the mix.
        + The beads contain complementary sequences to the adapters allowing each fragment to bind to a bead.
        + When the fragments attach to the beads the DNA denatures.
        + The fragments of DNA are then copied numerous times on each bead by PCR.
        + The beads are then filtered to remove any that without a fragment or contain more than one DNA fragment.
        + The remaining beads are put into wells on a sequencing plate (one bead per well) along with enzyme beads that contain the DNA polymerase and primer needed for the sequencing reaction.
        + The polymerase enzyme and primer attach to the DNA fragments on the beads.
        + Nucleotide bases are added to the wells one base at a time. When the base is incorporated, an H+ is released. 
        + The change in pH confirms the incorporation, and corresponds to the number of nucleotides of the same type that have been incorporated.
        + By plotting this change in pH, the sequence of the original piece of DNA can be decoded.
